# Μάθημα 4: Generative AI & Large Language Models (LLMs)

Φτάσαμε στην κορυφή της πυραμίδας: τα Transformers (τα μοντέλα πίσω από το ChatGPT). Εδώ δεν εκπαιδεύουμε μοντέλα από το μηδέν (χρειάζονται εκατομμύρια δολάρια για αυτό!). Κατεβάζουμε **έτοιμα** μοντέλα (Pre-trained) από τη **Hugging Face** και μαθαίνουμε πώς να τους 'μιλάμε' σωστά (Prompt Engineering).

## 1. Το Hugging Face Pipeline
Το pipeline είναι το πιο εύκολο εργαλείο για να κατεβάσεις ένα μοντέλο (εδώ θα χρησιμοποιήσουμε ένα πολύ μικρό μοντέλο της σειράς GPT, το `gpt2`, για να τρέξει γρήγορα σε κάθε υπολογιστή).

```python
from transformers import pipeline

# Κατεβάζουμε/Φορτώνουμε το μοντέλο
print("Φόρτωση του GPT-2...")
generator = pipeline('text-generation', model='gpt2')
print("Το μοντέλο είναι έτοιμο να παράγει κείμενο!")
```

## 2. Zero-Shot Prompting
Zero-Shot σημαίνει ότι ζητάμε από το μοντέλο να κάνει κάτι, *χωρίς* να του δώσουμε κανένα παράδειγμα.

```python
prompt_zero = "Translate the following English sentence to French: 'Hello, how are you?'"

# (Σημείωση: Το GPT-2 είναι παλιό μοντέλο και ίσως μπερδευτεί σε σύνθετα tasks. 
# Σε ένα πραγματικό project, θα βάζαμε το LLaMA-3 ή το Mistral εδώ!)
result_zero = generator(prompt_zero, max_new_tokens=15, num_return_sequences=1)

print("Zero-Shot Αποτέλεσμα:")
print(result_zero[0]['generated_text'])
```

## 3. Few-Shot Prompting (Η μαγεία του In-Context Learning)
Αν το μοντέλο δεν καταλαβαίνει τι θέλουμε, του δίνουμε 2-3 παραδείγματα μέσα στο Prompt. Το μοντέλο αναγνωρίζει το μοτίβο και συνεχίζει!

```python
prompt_few = """
Word: Happy
Antonym: Sad

Word: Hot
Antonym: Cold

Word: Fast
Antonym:"""

result_few = generator(prompt_few, max_new_tokens=5, num_return_sequences=1)

print("Few-Shot Αποτέλεσμα (Το μοντέλο κατάλαβε ότι ψάχνουμε αντίθετα!):")
print(result_few[0]['generated_text'])
```
