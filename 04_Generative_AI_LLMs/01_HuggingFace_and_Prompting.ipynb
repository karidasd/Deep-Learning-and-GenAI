## Χρήση του Hugging Face Pipeline
Η Hugging Face έχει φτιάξει ένα 'Pipeline' που μας επιτρέπει να κατεβάσουμε έτοιμα μοντέλα με 2 γραμμές κώδικα (Σημείωση: για μεγάλα μοντέλα απαιτείται Google Colab και GPU).

```python
from transformers import pipeline

# Κατεβάζουμε ένα μικρό μοντέλο παραγωγής κειμένου (GPT-2 για ταχύτητα)
print("Κατέβασμα του μοντέλου GPT-2...")
generator = pipeline('text-generation', model='gpt2')

# Του δίνουμε το αρχικό κείμενο (Prompt)
prompt = "The future of Artificial Intelligence is"
print(f"\nΑρχικό Κείμενο (Prompt): '{prompt}'")

# Παράγουμε τη συνέχεια...
result = generator(prompt, max_length=50, num_return_sequences=1, truncation=True)

print("\n--- Αποτέλεσμα (Generated Text) ---")
print(result[0]['generated_text'])
```
