Το PyTorch είναι το πιο δημοφιλές framework της Python για έρευνα σε AI. Όλα ξεκινάνε από ένα βασικό στοιχείο: τον **Τανυστή (Tensor)**.

## 1. Τι είναι το Tensor;
Το Tensor είναι απλά ένας πίνακας αριθμών (όπως τα arrays του NumPy), αλλά με μια υπερδύναμη: Μπορεί να 'τρέξει' σε Κάρτες Γραφικών (GPUs) για απίστευτα γρήγορους υπολογισμούς.

```python
import torch

# Δημιουργία ενός Tensor 2x3 με τυχαίους αριθμούς
x = torch.rand(2, 3)
print("Ο Τανυστής μου:\n", x)
print("\nΣχήμα (Shape):", x.shape)
```

## 2. Το πρώτο μας Νευρωνικό Δίκτυο (Multi-Layer Perceptron)
Θα φτιάξουμε ένα πολύ απλό δίκτυο που παίρνει μια είσοδο με 10 νούμερα, περνάει από ένα κρυφό επίπεδο (hidden layer) με 20 νευρώνες, και βγάζει 2 αποτελέσματα.

```python
import torch.nn as nn

class SimpleNetwork(nn.Module):
    def __init__(self):
        super(SimpleNetwork, self).__init__()
        # Επίπεδο 1: 10 είσοδοι -> 20 έξοδοι
        self.fc1 = nn.Linear(in_features=10, out_features=20)
        # Συνάρτηση Ενεργοποίησης (Activation Function) ReLU
        self.relu = nn.ReLU()
        # Επίπεδο 2: 20 είσοδοι -> 2 τελικές έξοδοι
        self.fc2 = nn.Linear(in_features=20, out_features=2)

    def forward(self, x):
        # Η 'πορεία' των δεδομένων μέσα στο δίκτυο
        x = self.fc1(x)
        x = self.relu(x)
        x = self.fc2(x)
        return x

# Αρχικοποίηση του δικτύου
model = SimpleNetwork()
print(model)
```

## 3. Forward Pass (Πέρασμα Δεδομένων)
Ας δώσουμε κάποια τυχαία δεδομένα στο δίκτυό μας για να δούμε τι θα βγάλει.

```python
# Έστω ότι έχουμε 5 δείγματα, το καθένα με 10 χαρακτηριστικά
random_input = torch.rand(5, 10)

# Περνάμε τα δεδομένα στο μοντέλο
output = model(random_input)

print("Αποτέλεσμα (5 δείγματα, 2 πιθανές απαντήσεις):\n", output)
```
