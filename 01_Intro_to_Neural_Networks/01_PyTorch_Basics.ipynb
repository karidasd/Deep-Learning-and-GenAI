# Μάθημα 1: Εισαγωγή στα Νευρωνικά Δίκτυα με PyTorch

Καλώς ήρθατε στον κόσμο του **Deep Learning**! 
Το PyTorch είναι το πιο δημοφιλές framework της Python για έρευνα σε AI. Όλα ξεκινάνε από ένα βασικό στοιχείο: τον **Τανυστή (Tensor)**.

## 1. Τι είναι το Tensor και το Autograd;
Το Tensor είναι ένας πίνακας αριθμών που μπορεί να 'τρέξει' σε GPUs. Η υπερδύναμή του όμως λέγεται **Autograd**: Μπορεί να υπολογίζει αυτόματα τις μαθηματικές παραγώγους (gradients), που είναι η καρδιά της μάθησης!

```python
import torch

# Δημιουργία ενός Tensor που απαιτεί παραγώγους (requires_grad=True)
x = torch.tensor([[1.0, 2.0], [3.0, 4.0]], requires_grad=True)

# Μια απλή μαθηματική πράξη
y = x.mean()

# Υπολογισμός των παραγώγων 'προς τα πίσω' (Backpropagation)
y.backward()

print("Ο Τανυστής μου:\n", x)
print("\nΟι αυτόματα υπολογισμένες παράγωγοι (Gradients):\n", x.grad)
```

## 2. Το πρώτο μας Νευρωνικό Δίκτυο (Multi-Layer Perceptron)
Θα φτιάξουμε ένα δίκτυο που παίρνει μια είσοδο με 10 νούμερα, περνάει από ένα κρυφό επίπεδο με 20 νευρώνες (με συνάρτηση ενεργοποίησης ReLU), και βγάζει 1 τελικό αποτέλεσμα.

```python
import torch.nn as nn
import torch.optim as optim

class SimpleNetwork(nn.Module):
    def __init__(self):
        super(SimpleNetwork, self).__init__()
        self.fc1 = nn.Linear(in_features=10, out_features=20)
        self.relu = nn.ReLU()
        self.fc2 = nn.Linear(in_features=20, out_features=1)

    def forward(self, x):
        x = self.fc1(x)
        x = self.relu(x)
        x = self.fc2(x)
        return x

model = SimpleNetwork()
print(model)
```

## 3. Η Διαδικασία Εκπαίδευσης (Training Loop)
Ένα Νευρωνικό δίκτυο στην αρχή είναι 'χαζό'. Πρέπει να του δείξουμε δεδομένα ξανά και ξανά (Epochs). 
Σε κάθε επανάληψη:
1. Κάνει μια πρόβλεψη (Forward Pass).
2. Βλέπει πόσο λάθος έκανε (Loss Function).
3. Διορθώνει τα βάρη του (Backward Pass & Optimizer).

```python
# Συνάρτηση κόστους (Mean Squared Error - Μετράει το λάθος)
criterion = nn.MSELoss()
# Ο Optimizer (SGD - Stochastic Gradient Descent) 'αναλαμβάνει' να διορθώσει το μοντέλο
optimizer = optim.SGD(model.parameters(), lr=0.01)

# Dummy δεδομένα (100 δείγματα, το καθένα με 10 χαρακτηριστικά)
X_train = torch.rand(100, 10)
# Dummy σωστοί στόχοι (τι θέλουμε να προβλέψει)
y_train = torch.rand(100, 1)

print("--- Έναρξη Εκπαίδευσης ---")
epochs = 50
for epoch in range(epochs):
    # 1. Forward Pass (Τι προβλέπει τώρα;)
    predictions = model(X_train)
    
    # 2. Υπολογισμός Λάθους
    loss = criterion(predictions, y_train)
    
    # 3. Καθαρισμός παλιών παραγώγων, Backpropagation και ενημέρωση βαρών
    optimizer.zero_grad() # Πάντα μηδενίζουμε τις προηγούμενες παραγώγους
    loss.backward()       # Υπολογισμός νέων παραγώγων
    optimizer.step()      # Αναβάθμιση του μοντέλου!
    
    if (epoch + 1) % 10 == 0:
        print(f"Epoch [{epoch+1}/{epochs}], Loss (Λάθος): {loss.item():.4f}")

print("Η εκπαίδευση ολοκληρώθηκε! Το λάθος μειώθηκε, άρα το μοντέλο έμαθε!")
```
